# **Data Reading**
 --Read Data DynanicallyUsing Parameter

# Create Stream To Read From Source


```python
dbutils.widgets.text("file_name","") #Empty String no Default Value
```

```python
p_file_name=dbutils.widgets.get("file_name")
```

```python
df = spark.readStream.format("cloudFiles")\
                      .option("cloudFiles.format", "csv")\
                      .option("cloudFiles.schemaLocation",f"abfss://bronze@databricksecommerce.dfs.core.windows.net/checkpoint_{p_file_name}")\
                      .load(f"abfss://source@databricksecommerce.dfs.core.windows.net/{p_file_name}")
 
```

# Create Stream To Read From Source to Bronze Layer

```python
df.writeStream.format("parquet")\
    .outputMode("append")\
    .option("checkpointLocation",f"abfss://bronze@databricksecommerce.dfs.core.windows.net/checkpoint_{p_file_name}")\
    .option("path",f"abfss://bronze@databricksecommerce.dfs.core.windows.net/{p_file_name}")\
    .trigger(once=True)\
    .start()   
```
